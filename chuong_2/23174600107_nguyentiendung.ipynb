## Nhập hai bảng student(student_id, name, class, course_id, score) và course(id, course_name)

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('dung.db')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS course (
    id INTEGER,
    course_name TEXT
)
""")

students = [
(1,"Nguyen Minh Hoang","May Tinh",12,6.7),
(2,"Tran Thi Lan","Kinh Te",34,9.2),
(3,"Pham Van Nam","Toan Tin",None,7.9),
(4,"Le Thanh Huyen","Toan Tin",20,7.2),
(5,"Vu Quoc Anh","May Tinh",24,8.0),
(6,"Dang Thuy Linh","May Tinh",24,5.5),
(7,"Bui Tien Dung","Kinh Te",34,9.2),
(8,"Ho Ngoc Mai","Toan Tin",20,8.8),
(9,"Duong Huu Phuc","Kinh Te",None,7.2),
(10,"Cao Thi Hanh","May Tinh",None,7.0)
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)",students)

courses = [
(12,"Giai tich"),
(34,"Thong ke"),
(26,"Tin hoc")
]

cursor.executemany("INSERT INTO course VALUES (?,?)",courses)

conn.commit()
```

## 1. Kết nối 2 bảng theo những cách:
### -Sử dụng tích Decartes.
### -Sử dụng JOIN: INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN

### Sử dụng tích Decartes:

```python
query = """
SELECT *
FROM student
CROSS JOIN course
"""
pd.read_sql(query,conn)
```

### INNER JOIN:

```python
query = """
SELECT *
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""
pd.read_sql(query,conn)
```

### LEFT JOIN:

```python
query = """
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""
pd.read_sql(query,conn)
```

### RIGHT JOIN:

```python
query = """
SELECT *
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
pd.read_sql(query,conn)
```

### FULL OUTER JOIN:

```python
query = """
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id

UNION

SELECT *
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
pd.read_sql(query,conn)
```

### Loại bỏ những bản ghi tham gia những môn học không tồn tại bảng course

```python
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (
SELECT id FROM course
)
""")

conn.commit()

pd.read_sql("SELECT * FROM student", conn)
```

### a. Tổng số sinh viên, điểm trung bình của từng lớp

```python
query = """
SELECT
class,
COUNT(*) AS total_students,
AVG(score) AS avg_score
FROM student
GROUP BY class
"""

pd.read_sql(query, conn)
```

### b. Tổng số sinh viên, điểm trung bình của từng môn học.

```python
query = """
SELECT
c.course_name,
COUNT(s.student_id) AS total_students,
AVG(s.score) AS avg_score
FROM student s
JOIN course c
ON s.course_id = c.id
GROUP BY c.course_name
"""

pd.read_sql(query, conn)
```

### c. Phân loại thi đua theo số điểm của từng môn học

```python
query = """
SELECT
name,
score,
CASE
WHEN score >= 9 THEN 'Xuat sac'
WHEN score >= 6 AND score <= 8.9 THEN 'Tot'
ELSE 'Kem'
END AS classification
FROM student
"""

pd.read_sql(query, conn)
```

## 4. Hãy bổ sung thêm một trường graduation_date có kiểu dữ liệu là DATETIME vào bảng student để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp được xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số.

### Thêm cột graduation_date

```python
cursor.execute("""
ALTER TABLE student
ADD COLUMN graduation_date DATETIME
""")

conn.commit()
```

### Tính thứ hạng theo điểm số

```python
query = """
SELECT
student_id,
name,
score,
RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
"""

pd.read_sql(query, conn)
```

```python
cursor.execute("""
UPDATE student
SET graduation_date =
datetime('now','+' || (
SELECT rank_score
FROM (
SELECT student_id,
RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
) r
WHERE r.student_id = student.student_id
) || ' days')
""")

conn.commit()
```
